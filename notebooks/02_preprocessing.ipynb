```python
import pandas as pd

nsl_train = pd.read_csv("/content/drive/MyDrive/nsl_kdd/nsl_kdd_train_processed.csv")
nsl_test = pd.read_csv("/content/drive/MyDrive/nsl_kdd/nsl_kdd_test_processed.csv")

categorical_cols = nsl_train.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols)

X_train = nsl_train.drop(columns=['attack_category'])
y_train = nsl_train['attack_category']

X_test = nsl_test.drop(columns=['attack_category'])
y_test = nsl_test['attack_category']

print(X_train.shape, y_train.shape)

categorical_features = ['protocol_type', 'service', 'flag']
numerical_features = [col for col in X_train.columns if col not in categorical_features]

print("Numerical features count:", len(numerical_features))
print("Categorical features:", categorical_features)

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)


X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

import numpy as np

np.save("/content/drive/MyDrive/nsl_kdd/X_train.npy", X_train_processed)
np.save("/content/drive/MyDrive/nsl_kdd/X_test.npy", X_test_processed)

np.save("/content/drive/MyDrive/nsl_kdd/y_train.npy", y_train_encoded)
np.save("/content/drive/MyDrive/nsl_kdd/y_test.npy", y_test_encoded)

import joblib

joblib.dump(preprocessor, "/content/drive/MyDrive/nsl_kdd/preprocessor.pkl")
joblib.dump(label_encoder, "/content/drive/MyDrive/nsl_kdd/label_encoder.pkl")

print("""
Phase 2 Summary:
- Numerical features scaled
- Categorical features one-hot encoded
- Target labels encoded
- Train and test sets transformed
Dataset is now ready for baseline models and deep learning implementation.
""")
```
